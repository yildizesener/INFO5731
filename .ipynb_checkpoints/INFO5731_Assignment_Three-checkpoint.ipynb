```python
from nltk.util import ngrams 
import nltk, re, string, collections
from nltk.corpus import stopwords
```

```python
with open('reviews.txt') as fin:
    lines = fin.readlines()

text = ' '.join([line.strip() for line in lines])
```

```python
# Text Preprocessing
numbers = '0123456789'
text = text.lower()
text = ''.join([char for char in text if char not in string.punctuation])
for ws in string.whitespace:
    text = text.replace(ws, ' ')
text = re.sub(r'\w*\d', '', text)
text = re.sub(' +', ' ', text)
```

```python
text[:100]
```

```python
tokens = nltk.word_tokenize(text)
stop_words = stopwords.words('english')
tokens = [token for token in tokens if token not in stop_words]
```

```python
#find n-grams
def find_n_grams(tokens, order):
    t = ngrams(tokens, order)
    t_freq = collections.Counter(t)
    return t_freq
```

```python
bigrams = find_n_grams(tokens, 2)
```

```python
trigrams = find_n_grams(tokens, 3)
```

```python
_ = [print(k, v) for k, v in trigrams.items()]
```
